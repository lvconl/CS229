```python
import numpy as np
import pandas as pd
from scipy.io import loadmat
```

```python
def load_weight(path):
    data = loadmat(path)
    return data['Theta1'], data['Theta2']
```

```python
theta1, theta2 = load_weight('datasets/ex3weights.mat')
theta1, theta2
```

```python
def load_data(path):
    data = loadmat(path)
    return data['X'], data['y']
```

```python
X, y = load_data('datasets/ex3data1.mat')
X, y
```

```python
def sigmoid_function(z):
    return 1 / (1 + np.exp(-z))
```

```python
X = np.insert(X, 0, values=np.ones(X.shape[0]), axis=1)
y = y.flatten()
```

```python
a1 = X
z2 = a1 @ theta1.T
```

```python
z2 = np.insert(z2, 0, 1, axis=1)
a2 = sigmoid_function(z2)
```

```python
z3 = a2 @ theta2.T
```

```python
a3 = sigmoid_function(z3)
```

```python
y_pred = np.argmax(a3, axis=1) + 1
acc = np.mean(y_pred == y)
print('acc is {}'.format(acc))
```
